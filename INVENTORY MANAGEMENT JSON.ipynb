```python
import json
product_id = str(input("Enter Product id:"))
name = str(input("Enter Name:"))
price = int(input("Enter Price:"))
quantity = int(input("Enter Quantity:"))
ingredients= str(input("Enter Ingredients:"))
weight= input("Enter Weight:")
benefits= str(input("Enter Benefits:"))
expiry_date = str(input("Enter expiry Date:"))
Inventory_Record[product_id]= {"name":name,"price":price,"quantity":quantity,"ingredients":ingredients,"weight":weight,
                "benefits":benefits,"expiry_date":expiry_date}
```

```python
js = json.dumps(Inventory_Record)
fd = open("Inventory_Record.json",'w')
fd.write(js)
fd.close()
```

```python
fd = open("Inventory_Record.json",'r')
r = fd.read()
fd.close()
Record=json.loads(r)
Record
```

```python
prod = str(input("Enter The product Id:"))
quant = int(input("Enter the quanity:"))
print("Product:", Inventory_Record[prod]['name'])
print("Price:", Inventory_Record[prod]['price'])
print("Ingredients:", Inventory_Record[prod]['ingredients'])
print("Weight:", Inventory_Record[prod]['weight'])
print("Benefits:", Inventory_Record[prod]['benefits'])
print("expiry_date:", Inventory_Record[prod]['expiry_date'])
print("Billing Amount:",Inventory_Record[prod]['price'] * quant)
Inventory_Record[prod]['quantity'] = Inventory_Record[prod]['quantity'] - quant
```
